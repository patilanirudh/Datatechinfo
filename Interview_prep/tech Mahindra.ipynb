```python
WITH FilteredSales AS (
    SELECT 
        Product_Category, 
        SUM(Quantity * Price) AS Total_Sales
    FROM 
        Sales
    WHERE 
        Date BETWEEN '2022-02-01' AND '2022-04-25'
    GROUP BY 
        Product_Category
)
SELECT 
    Product_Category, 
    Total_Sales
FROM 
    FilteredSales
WHERE 
    Total_Sales < 1000;
```
